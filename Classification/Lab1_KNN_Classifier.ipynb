```python
import pandas as pd

df=pd.read_csv('../data/fraud.csv', index_col = 0)

# Print the number of features, number of data points
print(df.shape)
#print(df.info())

# Print the name of the features of the data points
print(df.columns)

# Check your data and define X and y : What is X and what is y? 
y = df['Class'].values #target 
print("Y:",y)

df = df.iloc[:,1:]
print("1: ", df.shape)
print("1: column: ",df.columns)
X = df.drop(columns = 'Class').values #remaining columns
print("feature columns: ",X)
print("2: ", df.shape)
print("2: column: ",df.columns)
```

```python
# perform train - test split. (40% as testing and 60% as training)
from sklearn.model_selection import train_test_split

#X_train, X_test, y_train, y_test = train_test_split()

X_train, X_test, y_train, y_test = train_test_split(X,y, 
                                    test_size = 0.40, 
                                    random_state=1)
X_test.shape

print("X_test: ",X_test.shape);
print("X_train: ",X_train.shape);
print("Y_test: ",y_test.shape);
print("Y_train: ",y_train.shape);
```

```python
#--------------------------------------------------
## ----------- K-NN Classifier ------------------## 
#--------------------------------------------------
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier

steps = [('scaler', StandardScaler()),         
         ('knn', KNeighborsClassifier(n_neighbors = 3))]

knn_pipeline = Pipeline(steps)
knn_pipeline.fit(X_train, y_train)
```

```python
#--------------------------------------------------
## Model Evaluation ## 
#--------------------------------------------------
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score


ypred_test = knn_pipeline.predict(X_test)
df_test_result = pd.DataFrame({'Actual Test Data':y_test, 'Predict Text Data': ypred_test})
print(df_test_result)

mat_clf = confusion_matrix(y_test, ypred_test)
report_clf = classification_report(y_test, ypred_test)

print(mat_clf)
print(report_clf)

ypred_testP = knn_pipeline.predict_proba(X_test)
auc = roc_auc_score(y_test, ypred_testP[:,1])
print(auc)
```

```python
ypred_train = knn_pipeline.predict(X_train)

df_train_result = pd.DataFrame({'Actual':y_train, 'predict_values':ypred_train})
print(df_train_result)
print("");

mat_clf = confusion_matrix(y_train, ypred_train)
print("Confusion Matrix for train data: ")
print(mat_clf)
print("");

report_clf = classification_report(y_train, ypred_train)
print("Classification Report for train Data: ")
print(report_clf)
print("");

ypred_trainP = knn_pipeline.predict_proba(X_train)
auc = roc_auc_score(y_train, ypred_trainP[:,1])
print("ROC")
print(auc)
```
